**Figure 2**: Schematic of the isopycnal volume budget for the zonally-integrated Southern Ocean (SO). The volume below a given isopycnal surface ($S$) and poleward of latitude, $\phi$, is controlled by three general mechanisms: (1) The surface WMT ($G_{srf}$) arising from the surface buoyancy fluxes that leads to movement of the isopycnal layer interface, $S$, thus altering the volume of fluid within the layer; (2) Interior WMT ($G_{int}$), arising from the volume flux across $S$ by diapycnal mixing; and (3) The isopycnal overturning circulation at latitude $\phi$ ($\psi$), representing the volume transport across $\phi$ and below $S$. Our sign convention is such that $\psi < 0$ reflects the integrated transport of water away from the region south of $\phi$. $W$ represents the net volume gain/loss through the boundary due to precipitation, evaporation, runoff and sea ice and is less than 1\% of any of the other terms in the volume budget.
